```python
import cv2
import numpy as np
import os


if not os.path.exists("dataset"):
    os.makedirs("dataset")
    os.makedirs("dataset/train")
    os.makedirs("dataset/test")
    
    
    os.makedirs("dataset/train/0")
    os.makedirs("dataset/train/BOIRE")
    os.makedirs("dataset/train/JETAIME")
    os.makedirs("dataset/train/MERCI")
    os.makedirs("dataset/train/JOUER")
    os.makedirs("dataset/train/AMOUR")

    os.makedirs("dataset/test/0")
    os.makedirs("dataset/test/BOIRE")
    os.makedirs("dataset/test/JETAIME")
    os.makedirs("dataset/test/MERCI")
    os.makedirs("dataset/test/JOUER")
    os.makedirs("dataset/test/AMOUR")
##    os.makedirs("dataset/test/space")

# creation des dossiers de notre base des données
mode = 'train'
directory = 'dataset/'+mode+'/'

cap = cv2.VideoCapture(0)#ouvertur de camera

while True:
    _, frame = cap.read()#debut de capture

    frame = cv2.flip(frame, 1)#flip:pour la bonne suivi des main


    count = {
             'A': len(os.listdir(directory+"/AMOUR")),
             'B': len(os.listdir(directory+"/BOIRE")),
             'G': len(os.listdir(directory+"/JOUER")),
             'J': len(os.listdir(directory+"/JETAIME")),
             'M': len(os.listdir(directory+"/MERCI")),
             '0': len(os.listdir(directory+"/0")),
##             'space': len(os.listdir(directory+"/space")),

             }


    cv2.putText(frame, "MODE : "+mode, (10, 50), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "PROJET FIN D'ETUDE 21/22 :", (2, 20), cv2.FONT_HERSHEY_PLAIN, 0.8, (255,255,255), 1)
    cv2.putText(frame, "Nbr Images :", (10, 100), cv2.FONT_HERSHEY_COMPLEX, 0.7, (255,255,255), 1)
    cv2.putText(frame, "BOIRE : "+str(count['B']), (50, 135), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "JETAIME : "+str(count['J']), (50, 165), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "MERCI: "+str(count['M']), (50, 195), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "JOUER : "+str(count['G']), (50, 225), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "AMOUR : "+str(count['A']), (50, 255), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
    cv2.putText(frame, "0 : "+str(count['0']), (50, 280), cv2.FONT_HERSHEY_PLAIN, 1, (0,255,255), 1)
   #affiche des mots de notre base de données avec leurs nombre



    #ajuster la fenetre de capture
    x1 = int(0.3*frame.shape[1])
    y1 = 5
    x2 = frame.shape[1]-10
    y2 = int(0.7*frame.shape[1])

    cv2.rectangle(frame, (x1-1, y1-1), (x2+1, y2+1), (255,0,0) ,1)

    roi = frame[y1:y2, x1:x2]
    roi = cv2.resize(roi, (400, 400))

    cv2.imshow("Capture", frame)

    #convertir les images en niveau e gris
    roi = cv2.cvtColor(roi, cv2.COLOR_BGR2RGB)
    _, roi = cv2.threshold(roi, 120, 255, cv2.THRESH_BINARY)#seuillage au niveau binaire
    cv2.imshow("Image", roi)

    interrupt = cv2.waitKey(10)
    if interrupt & 0xFF == 27: # esc key
        break
    
        
    if interrupt & 0xFF == ord('A'):
        cv2.imwrite(directory+'AMOUR/'+str(count['A'])+'.jpg', roi)
    if interrupt & 0xFF == ord('B'):
        cv2.imwrite(directory+'BOIRE/'+str(count['B'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('G'):
        cv2.imwrite(directory+'JOUER/'+str(count['G'])+'.jpg', roi)
   
    if interrupt & 0xFF == ord('J'):
        cv2.imwrite(directory+'JETAIME/'+str(count['J'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('M'):
        cv2.imwrite(directory+'MERCI/'+str(count['M'])+'.jpg', roi)
    
    if interrupt & 0xFF == ord('0'):
        cv2.imwrite(directory+'0/'+str(count['0'])+'.jpg', roi)




cap.release()
cv2.destroyAllWindows()

```
